```python
df = df.iloc[1:]
df
```

### The first row has been converted to column names.

```python
df = df.loc[:, df.columns.notna()]
df
```

### Removed the NaN Column

```python
#df[' noise.dB'] = df[' noise.dB'].str.replace("dB","")
```

```python
df['price'] = df['price'].str.replace("approx","")
df
```

```python
df['price'] = df['price'].str.replace("$","")
df['price'] = df['price'].str.replace("open to negotiate","")
df['price'] = df['price'].str.replace("deal price","")
df['price'] = df['price'].str.replace("/- only","")
df['price'] = df['price'].str.replace(",","")
df
```

```python
df['price'] = df['price'].astype(int)
```

### Converted the prices to integers

```python
df['noise.dB.'] = df['noise.dB.'].str.replace("dB","")
```

```python
df['noise.dB.'] = df['noise.dB.'].astype(float)
df['noise.dB.'] = df['noise.dB.'].round()
df['noise.dB.'] = df['noise.dB.'].astype(int)
df

```
